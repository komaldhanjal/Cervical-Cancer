## checkiing Relatonship between independent and ddependent 

```python
for col in data[:-1]:
    plt.scatter(x = data[col], y = data["Biopsy"])
    plt.title(f"Relationship Between {col} and Biopsy")
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import  RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
x=data.iloc[:,:-1]
y=data["Biopsy"]
```

```python
X_train,X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
model = RandomForestClassifier(n_estimators = 100, random_state = 42, criterion = "gini", class_weight = "balanced")
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
evaluation=pd.DataFrame({
    "Actual Value":y_test,
    "Predicted Value":y_pred,
    "Error":y_test-y_pred})
evaluation
```

```python
accuracy = accuracy_score(y_test, y_pred) * 100
print(f"Model Accuracy is {accuracy}")
```

```python
confusion_m=confusion_matrix(y_test,y_pred)
```

```python
plt.figure(figsize = (10,8))
sns.heatmap(confusion_m, annot = True, fmt = "d", color = "blue")
plt.title("Confusion Metrix Heatmap")
plt.show()
```
